## 9.6	CONSULTAS COM JUNÇÃO E ORDENAÇÃO (Mínimo 6)

### a) Uma junção que envolva todas as tabelas possuindo no mínimo 3 registros no resultado

```python
resultx = pd.read_sql_query("""
                            select * from pessoa pe 
                            inner join usuario us on (pe.cod_pessoa = us.fk_pessoa_cod_pessoa)
                            inner join cidade ci on (ci.cod_cidade = pe.fk_cidade_cod_cidade)
                            inner join estado es on (es.cod_estado = ci.fk_estado_cod_estado)
                            inner join pais pa on (pa.cod_pais = es.fk_pais_cod_pais)
                            inner join postagem_posta_comentario ppc 
                            on (us.fk_pessoa_cod_pessoa = ppc.fk_usuario_fk_pessoa_cod_pessoa)
                            inner join livro lv on (us.fk_pessoa_cod_pessoa = lv.cod_livro)
                            inner join livro_genero lg on (lv.cod_livro = lg.fk_livro_cod_livro)
                            inner join genero ge on (lg.fk_genero_cod_genero = ge.cod_genero)
                            inner join comparece_evento ce on (us.fk_pessoa_cod_pessoa = ce.fk_usuario_fk_pessoa_cod_pessoa)
                            inner join evento ev on (ce.fk_evento_cod_evento = ev.cod_evento)
                            inner join evento_possui_tipo ept on (ev.cod_evento = ept.fk_evento_cod_evento)
                            inner join tipo_evento te on (ept.fk_tipo_evento_cod_tipo = te.cod_tipo)
                            """,conn)
```

```python
resultx
```
